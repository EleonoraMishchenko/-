## Тест на знание общих сведений о языке программирования Python (10 вопросов).

Программа выводит в консоль текст вопросов по одному и ждет ввода ответа от пользователя.
После ввода ответа пользователя программа выводит в консоль результат. 
Если пользователь дал правильный ответ, программа возвращает "Ответ {ОТВЕТ_ПОЛЬЗОВАТЕЛЯ} верен". 
Если пользователь ввел неправильный ответ, программа возвращает "Неверный ответ." и задает вопрос снова, пока пользователь не введет правильный ответ.

В конце теста программа пишет, сколько правильных и неправильных ответов дал пользователь.

Программа не учитывает регистр ответа пользователя.

В программу можно подставить любые вопросы.

```python
questions = {'В каком году вышла первая версия Python?': '1991',
             'Какая версия языка Python актуальна сейчас? Укажите цифрой.': '3',
             'Относятся ли к базовым типам данных в языке Python дробные числа (с плавающей точкой)?': 'да',
             'Укажите название специального типа данных, обозначающего "ничего".': 'none',
             'Может ли строка (str) быть пустой?': 'да',
             'Допускается ли в Python математическое вычитание строк?': 'нет',
             'Будет ли работать код с условием "if" без указания оператора "else"?': 'да',
             'Как правильно называется соединение 2 строк в единую строку?': 'конкатенация',
             'Какое значение будет присвоено переменной "a" при выполнении следующего кода:\nmy_string = "abcde"\na = my_string[-1]': 'e',
             'Какая кодировка используется в строках?': 'utf8'}

correct_answers = 0
incorrect_answers = 0
print('Вся наша жизнь игра! Приветствуем на тестовом задании на знание основ Python.\nВам будет предложено 10 вопросов, и только ответив на все вопросы правильно Вы сможете завершить тест.\nПриступим!')
for question, answer in questions.items():
    user_answer = input(question + "\n").strip().lower()
    while user_answer != answer.lower():
        print("Неверный ответ.")
        user_answer = input(question + "\n").strip().lower()
        incorrect_answers += 1
    print(f"Ответ {user_answer} верен.")
    correct_answers += 1

print(f"Количество правильных ответов: {correct_answers}")
print(f"Количество неправильных ответов: {incorrect_answers}")
```
